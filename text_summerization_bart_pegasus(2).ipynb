```python
text = """The brief facts of the case are that FIR dated 16.8.2007 was registered against 24 accused persons
including the appellant under sections 79 and 80 of the 1963 Act as they were found to be indulging 
in gambling.The charge sheet was filed and the Trial Court vide order 
dated 21.8.2007 convicted them under Section 79 & 80 of the 1963 Act 
and sentenced them to undergo imprisonment for a period of one year 
each under both the provisions along with a fine of ₹ 600/- after the 
accused had pleaded guilty. At that stage, the accused filed affidavits"""
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model = "google/pegasus-xsum")
summarizer(text)
```

```python
from transformers import pipeline
summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
summarizer(text)
```

```python
# automatically retrieve the relevant model given the name to the pretrained config

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

tokens_input = tokenizer.encode("summarize: "+text, return_tensors='pt', max_length=129, truncation=True)
ids = model.generate(tokens_input, min_length=20, max_length=50)
summary = tokenizer.decode(ids[0], skip_special_tokens=True)

print(summary)
```

```python
import gradio as gr
```

```python
def predict(text):
    summary = summarizer(text)
    return summary

gr.Interface(
    predict,
    inputs=gr.inputs.Textbox(placeholder="Text here..."),
    outputs="text",
    title="Summarizer",
).launch(debug=False)
```

```python
from transformers import pipeline
pipeline = pipeline(task="summarization", model = "facebook/bart-large-cnn")
```

```python
def predict(text):
    predictions = pipeline(text)
    return predictions

gr.Interface(
    predict,
    inputs=gr.inputs.Textbox(label="Input Text", type="text"),
    outputs=gr.outputs.Textbox(label='Ouput Text', type="text"),
    title="Text Summarizer",
).launch()
```
