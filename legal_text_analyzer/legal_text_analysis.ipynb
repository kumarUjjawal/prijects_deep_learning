```python
# importing required modules
from PyPDF2 import PdfReader

# creating a pdf reader object
reader = PdfReader('15-May-2023.pdf')

# printing number of pages in pdf file
print(len(reader.pages))
pagelenght = len(reader.pages)
# getting a specific page from the pdf file
page = reader.pages[pagelenght-1]

# extracting text from page
text = page.extract_text()
print(text)
```

```python
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LAParams, LTFigure, LTTextBox
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFPageInterpreter, PDFResourceManager
from pdfminer.pdfpage import PDFPage, PDFTextExtractionNotAllowed
from pdfminer.pdfparser import PDFParser

text = ""

with open('15-May-2023.pdf', 'rb') as f:
    parser = PDFParser(f)
    doc = PDFDocument(parser)
    page = list(PDFPage.create_pages(doc))[0]
    rsrcmgr = PDFResourceManager()
    device = PDFPageAggregator(rsrcmgr, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    interpreter.process_page(page)
    layout = device.get_result()
    
    for obj in layout:
        if isinstance(obj, LTTextBox):
            text += obj.get_text()
```

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_sm')

pattern = [{'LOWER': 'section'}, {'IS_DIGIT': True},]

matcher = Matcher(nlp.vocab)
matcher.add('SECTION_PATTERN', [pattern])

text = "According to Section 70 of the law, all citizens have the right to vote."

doc = nlp(text)

matches = matcher(doc)

for match_id, start, end in matches:
    matched_span = doc[start:end]
    print(matched_span.text)
```
